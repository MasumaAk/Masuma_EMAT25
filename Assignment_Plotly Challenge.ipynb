# Assignment of Plotly

```python
import plotly.express as px
import pandas as pd
import json
```

#### Importing the dataset

```python
gdh = pd.read_csv('archive.csv')
```

#### Create two figures using the data frame and display them. Use px.bar.

```python
fig1 = px.bar(gdh, x = 'Year', y= 'February Average Temperature')
fig2 = px.bar(gdh, x = 'Year', y= 'March Average Temperature')

fig1.show()
fig2.show()
```

#### Using a color discrete map, add designated colors to each category within the data

```python
fig3 = px.bar(gdh, x = 'Year', y= 'February Average Temperature', color= 'Punxsutawney Phil',
             color_discrete_map={'No Shadow': 'lightblue', 'Full Shadow': 'lightgreen', 'No Record': 'orange'},)
fig4 = px.bar(gdh, x = 'Year', y= 'March Average Temperature', color= 'Punxsutawney Phil',
             color_discrete_map={'No Shadow': 'lightblue', 'Full Shadow': 'lightgreen', 'No Record': 'orange'},)


fig3.show()
fig4.show()
```

#### Using category_orders to categorize them by year

```python
fig5 = px.bar(gdh, x = 'Year', y= 'February Average Temperature', color= 'Punxsutawney Phil',
             color_discrete_map={'No Shadow': 'lightblue', 'Full Shadow': 'lightgreen', 'No Record': 'orange'},
             category_orders= {'Year': sorted(gdh['Year'].unique())})

fig6 = px.bar(gdh, x = 'Year', y= 'March Average Temperature', color= 'Punxsutawney Phil',
             color_discrete_map={'No Shadow': 'lightblue', 'Full Shadow': 'lightgreen', 'No Record': 'orange'},
             category_orders= {'Year': sorted(gdh['Year'].unique())})


fig5.show()
fig6.show()
```

```python
#An alternative option
color_map = {
    'No Shadow': 'lightblue',
    'Full Shadow': 'lightgreen',
    'No Record': 'orange',
    'Partial Shadow': 'purple'  # Assign a unique color
}

fig5a = px.bar(gdh, x='Year', y='February Average Temperature', color='Punxsutawney Phil',
              color_discrete_map=color_map)

fig6a = px.bar(gdh, x='Year', y='March Average Temperature', color='Punxsutawney Phil',
              color_discrete_map=color_map)

fig5a.show()
fig6a.show()

```

#### Use fig.updatelayout to  update the labels, color, and title of the grap

```python
fig7 = px.bar(gdh, x = 'Year', y = 'February Average Temperature', color = 'Punxsutawney Phil',
              color_discrete_map = {'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
              category_orders = {'Year': sorted(gdh['Year'].unique())})
fig8 = px.bar(gdh, x = 'Year', y = 'March Average Temperature', color = 'Punxsutawney Phil',
              color_discrete_map = {'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
              category_orders = {'Year': sorted(gdh['Year'].unique())})

fig7.update_layout(
    title = 'Groundhog Feb',
    yaxis_title = 'Average Temp',
    xaxis_title = 'Year'
)

fig8.update_layout(
    title = 'Groundhog March',
    yaxis_title = 'Average Temp',
    xaxis_title = 'Year'
)

fig7.show()
fig8.show()
```

#### Enable scroll zoom

```python
fig9 = px.bar(gdh, x = 'Year', y = 'February Average Temperature', color = 'Punxsutawney Phil',
              color_discrete_map = {'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
              category_orders = {'Year': sorted(gdh['Year'].unique())})
fig10 = px.bar(gdh, x = 'Year', y = 'March Average Temperature', color = 'Punxsutawney Phil',
              color_discrete_map = {'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
              category_orders = {'Year': sorted(gdh['Year'].unique())})

fig9.update_layout(
    title = 'Groundhog Feb',
    yaxis_title = 'Average Temp',
    xaxis_title = 'Year'
)

fig10.update_layout(
    title = 'Groundhog March',
    yaxis_title = 'Average Temp',
    xaxis_title = 'Year'
)

config = {'scrollZoom': True}

fig9.show(config = config)
fig10.show()
```

#### Use px.scatter to make the histograms into scatter charts

```python


fig11 = px.scatter(gdh, x='Year', y='February Average Temperature', color='Punxsutawney Phil',
                   color_discrete_map={'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
                   category_orders={'Year': sorted(gdh['Year'].unique())})

fig12 = px.scatter(gdh, x='Year', y='March Average Temperature', color='Punxsutawney Phil',
                   color_discrete_map={'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
                   category_orders={'Year': sorted(gdh['Year'].unique())})

# Update layout for February plot
fig11.update_layout(
    title='Groundhog Feb',
    yaxis_title='Average Temp',
    xaxis_title='Year'
)

# Update layout for March plot
fig12.update_layout(
    title='Groundhog March',
    yaxis_title='Average Temp',
    xaxis_title='Year'
)

# Enable scroll zoom for both plots
fig11.show(config={'scrollZoom': True})
fig12.show(config={'scrollZoom': True})
```

#### Use update_layout to create zoom-in thumbnails of the histograms that can display certain data while still being interacted with.

```python
fig13 = px.histogram(gdh, x = 'Year', y = 'February Average Temperature', color = 'Punxsutawney Phil',
              color_discrete_map = {'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
              category_orders = {'Year': sorted(gdh['Year'].unique())})
fig14 = px.histogram(gdh, x = 'Year', y = 'March Average Temperature', color = 'Punxsutawney Phil',
              color_discrete_map = {'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
              category_orders = {'Year': sorted(gdh['Year'].unique())})

fig13.update_layout(
    title = 'Groundhog Feb',
    yaxis_title = 'Average Temp',
    xaxis_title = 'Year',
    xaxis = {'range': [100, 140]},
    yaxis = {'range': [20, 50]}
)

fig14.update_layout(
    title = 'Groundhog March',
    yaxis_title = 'Average Temp',
    xaxis_title = 'Year',
    xaxis = {'range': [60, 70]},
    yaxis = {'range': [0, 50]}
)

config = {'scrollZoom': True}

fig13.show(config = config)
fig14.show() 
```

```python
# Create the February Temperature Histogram
fig15 = px.histogram(gdh, x='Year', y='February Average Temperature', color='Punxsutawney Phil',
                     color_discrete_map={'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
                     category_orders={'Year': sorted(gdh['Year'].unique())})

# Create the March Temperature Histogram
fig16 = px.histogram(gdh, x='Year', y='March Average Temperature', color='Punxsutawney Phil',
                     color_discrete_map={'No Shadow': 'lightgreen', 'Full Shadow': 'lightblue', 'No Record': 'grey'},
                     category_orders={'Year': sorted(gdh['Year'].unique())})

# Update layout for February histogram with zoom-in thumbnails
fig15.update_layout(
    title='Groundhog Day - February Average Temperature',
    yaxis_title='Average Temperature (°F)',
    xaxis_title='Year',
    xaxis=dict(rangeslider=dict(visible=True), range=[100, 140]),  # Zoom-in range with slider
    yaxis=dict(range=[20, 50], rangemode="normal")  # Allow interaction while zoomed in
)

# Update layout for March histogram with zoom-in thumbnails
fig16.update_layout(
    title='Groundhog Day - March Average Temperature',
    yaxis_title='Average Temperature (°F)',
    xaxis_title='Year',
    xaxis=dict(rangeslider=dict(visible=True), range=[60, 70]),  # Zoom-in range with slider
    yaxis=dict(range=[0, 50], rangemode="normal")  # Keep interaction enabled
)

# Enable scroll zoom
config = {'scrollZoom': True}

# Show figures with zoom enabled
fig15.show(config=config)
fig16.show(config=config)
```
